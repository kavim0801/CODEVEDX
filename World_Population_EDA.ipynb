```python
df = pd.read_csv("World_population_CSV.csv")
```

```python
print("Mean:")
print(df.select_dtypes(include="number").mean())

print("\nMedian:")
print(df.select_dtypes(include="number").median())

print("\nMode:")
print(df.select_dtypes(include="number").mode().iloc[0])
```

```python
print("Standard Deviation:")
print(df.select_dtypes(include="number").std())
```

```python
continent_population = df.groupby("Continent")["Population"].mean().sort_values(ascending=False)

print(continent_population)
```

```python
country_population = df.groupby("Country")["Population"].mean().sort_values(ascending=False)

print(country_population)
```

```python
print(df.columns.tolist())
```

```python
# Find the population growth column
growth_column = [col for col in df.columns if "Growth" in col or "growth" in col][0]

# Calculate average population growth by continent
growth_by_continent = (
    df.groupby("Continent")[growth_column]
    .mean()
    .sort_values(ascending=False)
)

print(growth_by_continent)
```

```python
plt.figure(figsize=(10, 6))

growth_by_continent.plot(kind="bar")

plt.title("Average Population Growth Rate by Continent")
plt.xlabel("Continent")
plt.ylabel("Average Population Growth (%)")

plt.xticks(rotation=45)
plt.show()
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(10, 6))

plt.hist(df["Population"], bins=10)

plt.title("Distribution of World Population")
plt.xlabel("Population")
plt.ylabel("Frequency")

plt.show()
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(10, 6))

plt.boxplot(df["Population"].dropna())

plt.title("Box Plot of World Population")
plt.ylabel("Population")

plt.show()
```

```python
Q1 = df["Population"].quantile(0.25)
Q3 = df["Population"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

population_outliers = df[
    (df["Population"] < lower_bound) |
    (df["Population"] > upper_bound)
]

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

print("\nPopulation Outliers:")
print(population_outliers[["Country", "Population"]])
```

```python
correlation = df.select_dtypes(include="number").corr()

print(correlation)
```

```python
plt.figure(figsize=(10, 6))

sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Heatmap of World Population Data")

plt.show()
```

# 🌍 World Population Exploratory Data Analysis (EDA)

## Project Overview

This project performs Exploratory Data Analysis (EDA) on a World Population dataset using Python.

The purpose of this analysis is to understand population patterns through statistical summaries, comparative analysis, outlier detection, and data visualization.

### Objectives

- Understand the structure of the dataset.
- Calculate descriptive statistics.
- Compare population patterns.
- Detect potential outliers.
- Visualize population distributions.
- Analyze correlations between numerical variables.
- Interpret the findings and draw meaningful conclusions.

- Python
- Pandas
- NumPy

- Seaborn
- Jupyter Notebook
